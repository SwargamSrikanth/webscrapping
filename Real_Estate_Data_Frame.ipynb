```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
url="https://www.makaan.com/visakhapatnam-residential-property/buy-property-in-visakhapatnam-city?_=1649941405596"
```

```python
list_bhk=[]
l_bhk=[]
location=[]
list_price=[]
list_sq_ft=[]
list_area=[]
list_status=[]
list_possession=[]
builder_list=[]
RERA=[]
list_seller=[]
url="https://www.makaan.com/visakhapatnam-residential-property/buy-property-in-visakhapatnam-city?_=1649941405596"
for i in range(1,150):
    print(url+"&page="+str(i))
    urls = url+"&page="+str(i)
    p=requests.get(urls)
    p.text
    soup=bs(p.text)
    
    #creating llist of bhks
    bhk=soup.find_all("div",class_='title-line-wrap')
    for j in bhk:
        list_bhk.append(j.text)
        
    loc=soup.find_all("span",class_="locName")
    for j in loc:
        location.append(j.text)

    price=soup.find_all("td",class_='price')
    for j in price:
        list_price.append(j.text)
    
    sq_ft=soup.find_all("td",class_="lbl rate")
    for j in sq_ft:
        list_sq_ft.append(j.text)
        
    area=soup.find_all('td',class_="size")
    for j in range(len(area)):
        list_area.append(area[j].text)
    
    
    status=soup.find_all("td",class_="val")
    for j in status:
        list_status.append(j.text)
    
    possession=soup.find_all("ul",class_="listing-details")
    for j in range(len(possession)):
        list_possession.append(possession[j].text)
    
    for i in soup.find_all('div',class_='seller-info'):
        info=i.find('a',class_='seller-name')
        if info :
            list_seller.append(info.text)
        else:
            list_seller.append(np.nan)
            
    for i in soup.find_all('div',class_='title-line'):
        info=i.find('div',class_='rera-tag-new')
        if info is None:
            RERA.append('No')
        else:
            RERA.append('Yes')
    
        

print(len(list_area))
print(len(list_price))
print(len(list_sq_ft))
print(len(list_status))
print(len(list_bhk))
print(len(list_possession))
print(len(location))
print(len(RERA))
print(len(list_seller))
```

```python
 x={"proprty_type":list_bhk,"seller":list_seller,"Location":location,"Area in sqft":list_area,"Rate/sq_ft":list_sq_ft,"price":list_price,"status":list_status,"possession":list_possession,"Approved by RERA":RERA}
```
